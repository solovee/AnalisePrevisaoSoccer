```python
import os
import pandas as pd

lista_dfs = []

for file in os.listdir():
    if file.startswith('premier_league') and file.endswith('.csv'):
        try:
            df_temp = pd.read_csv(file, sep=None, engine='python', encoding='utf-8', on_bad_lines='skip')
        except UnicodeDecodeError:
            try:
                df_temp = pd.read_csv(file, sep=None, engine='python', encoding='latin1', on_bad_lines='skip')
            except Exception as e:
                print(f"Erro ao ler {file}: {e}")
                continue
        
        lista_dfs.append(df_temp)
        print(f"Lido com sucesso: {file}")

df_pl = pd.concat(lista_dfs, ignore_index=True, join='outer')
```

```python
nans = df_pl.isna().sum()
```

```python
nans[15:30]
```

```python
df_pl.to_csv('premier_league_all.csv', index=False, encoding='utf-8')
```
