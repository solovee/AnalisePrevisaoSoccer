```python
import pandas as pd
import os

file_name = "matches_2023.csv"

for league in lista_ids[500:]:
    print(f'Coletando dados da liga {league}')
    a = api_obj.get_fixtures(league_id=league, season=2023)

    data = []
    for match in a['response']:
        data.append({
            'fixture_id': match["fixture"]["id"],
            "home_id": match["teams"]["home"]["id"],
            "home_name": match["teams"]["home"]["name"],
            "away_id": match["teams"]["away"]["id"],
            "away_name": match["teams"]["away"]["name"],
            "score_home": match["goals"]["home"],
            "score_away": match["goals"]["away"],
            "timestamp": match["fixture"]["timestamp"],
            "venue_id": match["fixture"]["venue"]["id"],
            "venue": match["fixture"]["venue"]["name"],
            "venue_city": match["fixture"]["venue"]["city"],
            "league": match["league"]["name"],
            "league_id": match["league"]["id"],
            "league_country": match["league"]["country"],
            "winner_home": match["teams"]["home"]["winner"],
            "winner_away": match["teams"]["away"]["winner"]
        })

    # cria dataframe da liga atual
    df = pd.DataFrame(data)

    # salva no csv (append se já existir, cria se não existir)
    if not os.path.isfile(file_name):
        df.to_csv(file_name, index=False, mode="w", header=True)
    else:
        df.to_csv(file_name, index=False, mode="a", header=False)
```

```python
lista_ids_23 = df['fixture_id'].to_list()
```

```python
import pandas as pd
import time

dados = []

for i, fixture in enumerate(lista_ids_23, start=1):
    stats = api_obj.get_fixture_statistics(fixture_id=fixture)
    res = stats.get('response', [])

    if not res:
        continue

    for team_data in res:
        team_info = team_data.get('team', {})
        stats_list = team_data.get('statistics', [])
        
        stats_dict = {s['type']: s['value'] for s in stats_list}
        stats_dict['fixture_id'] = fixture
        stats_dict['team_id'] = team_info.get('id')
        stats_dict['team_name'] = team_info.get('name')

        dados.append(stats_dict)
    
    # A cada 100 fixtures, salva o progresso
    if i % 100 == 0 or i == len(lista_ids_23):
        df_temp = pd.DataFrame(dados)
        cols = ['fixture_id', 'team_id', 'team_name'] + [c for c in df_temp.columns if c not in ['fixture_id', 'team_id', 'team_name']]
        df_temp = df_temp[cols]
        
        df_temp.to_csv(f'stats_progress_{i}.csv', index=False)
        print(f'Salvo progresso após {i} fixtures -> stats_progress_{i}.csv')

        # Pausa leve opcional para não sobrecarregar a API
        time.sleep(1)

print("✅ Coleta completa!")
```
